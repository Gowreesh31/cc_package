# Smart Supply Chain Forecasting
This notebook trains a Prophet model to predict stock levels.

```python
import pandas as pd
from prophet import Prophet
import boto3

# Load data from Athena/S3
s3 = boto3.client('s3')
obj = s3.get_object(Bucket='YOUR_CURATED_BUCKET', Key='daily_demand/part-00000.parquet')
df = pd.read_parquet(obj['Body'])

# Prepare for Prophet
df_prophet = df[df['sku'] == 'SKU-001'][['date', 'daily_demand']]
df_prophet.columns = ['ds', 'y']

# Train Model
m = Prophet()
m.fit(df_prophet)

# Forecast 30 days
future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)

print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```
